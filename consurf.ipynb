There are 6 running modes.

1) PDB: In this mode the user submits a model file in either mmCIF or PDB format. The user can upload the file or type a PDB/UNIPROT ID and colab will fetch the file. The UNIPROT file comes from alphafold.

2) PDB_MSA: The user provides both a model file and a MSA file. The user will be asked to choose the query sequence from the MSA. The calculation is done on this sequence.

3) PDB_MSA_TREE: The user provides a model file, a MSA file, and a tree file.

4) MSA: The user submits a MSA file.

5) MSA_TREE: The user provides a MSA file and a tree file.

6) Sequence: The user submits a sequence.

After choosing the mode, press the play button under 'Run ConSurf' (If you don't see the play button, move the cursor over 'Run ConSurf'). The first run is a little longer because colab installs the programs. After installing, the user will be asked if they would like to change the default parameters.
